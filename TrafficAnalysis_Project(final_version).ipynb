```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("/content/traffic.csv")

# Convert 'Time' column to datetime format for better analysis
df["Time"] = pd.to_datetime(df["Time"], format="%I:%M:%S %p").dt.hour  # Extract only the hour

# Ensure 'Traffic Situation' is categorical for proper visualization
df["Traffic Situation"] = df["Traffic Situation"].str.lower().astype("category")

# Set figure size
plt.figure(figsize=(14, 6))

# Task 1 Vehicle Distribution Analysis:

# Boxplot for vehicle counts
plt.subplot(1, 2, 1)
sns.boxplot(data=df[["CarCount", "BikeCount", "BusCount", "TruckCount"]])
plt.title("Boxplot of Vehicle Counts")
plt.ylabel("Count")


```

```python
# Histogram for vehicle counts
df[["CarCount", "BikeCount", "BusCount", "TruckCount"]].plot.hist(alpha=0.5, bins=20, figsize=(14, 6))
plt.title("Histogram of Vehicle Counts")
plt.xlabel("Count")

plt.show()
```

```python
# task 2: Traffic Situation Distribution:
# Bar chart
plt.figure(figsize=(12, 5))
sns.countplot(x=df["Traffic Situation"], order=df["Traffic Situation"].value_counts().index, palette="viridis")
plt.title("Distribution of Traffic Situations")
plt.xlabel("Traffic Situation")
plt.ylabel("Count")
plt.show()

# Pie chart
df["Traffic Situation"].value_counts().plot.pie(autopct="%1.1f%%", figsize=(8, 8), colors=["red", "orange", "yellow", "green"])
plt.title("Traffic Situation Distribution")
plt.ylabel("")
plt.show()
```

```python
# task 3: Variation by Day of the Week:
plt.figure(figsize=(14, 8))

# Bar plot for each vehicle type
plt.subplot(2, 2, 1)
sns.barplot(x="Day of the week", y="CarCount", data=df, estimator=sum, ci=None, palette="Blues")
plt.xticks(rotation=45)
plt.title("Car Count by Day of the Week")
plt.ylabel("Car Count")

plt.subplot(2, 2, 2)
sns.barplot(x="Day of the week", y="BikeCount", data=df, estimator=sum, ci=None, palette="Greens")
plt.xticks(rotation=45)
plt.title("Bike Count by Day of the Week")
plt.ylabel("Bike Count")

plt.subplot(2, 2, 3)
sns.barplot(x="Day of the week", y="BusCount", data=df, estimator=sum, ci=None, palette="Oranges")
plt.xticks(rotation=45)
plt.title("Bus Count by Day of the Week")
plt.ylabel("Bus Count")

plt.subplot(2, 2, 4)
sns.barplot(x="Day of the week", y="TruckCount", data=df, estimator=sum, ci=None, palette="Reds")
plt.xticks(rotation=45)
plt.title("Truck Count by Day of the Week")
plt.ylabel("Truck Count")

plt.tight_layout()
plt.show()
```

```python
# 4-7. Vehicle Counts vs. Traffic Situation (Scatter, Box, Violin Plots)
# Define function for visualization
def plot_vehicle_vs_traffic(vehicle_column, title):
    plt.figure(figsize=(12, 5))

    # Scatter plot
    plt.subplot(1, 3, 1)
    sns.scatterplot(x=df[vehicle_column], y=df["Traffic Situation"], hue=df["Traffic Situation"], palette="coolwarm")
    plt.title(f"{title} vs. Traffic Situation")

    # Box plot
    plt.subplot(1, 3, 2)
    sns.boxplot(x=df["Traffic Situation"], y=df[vehicle_column], palette="coolwarm")
    plt.title(f"{title} vs. Traffic Situation")

    # Violin plot
    plt.subplot(1, 3, 3)
    sns.violinplot(x=df["Traffic Situation"], y=df[vehicle_column], palette="coolwarm")
    plt.title(f"{title} vs. Traffic Situation")

    plt.show()

# Call function for each vehicle type
plot_vehicle_vs_traffic("CarCount", "Car Count")
plot_vehicle_vs_traffic("BikeCount", "Bike Count")
plot_vehicle_vs_traffic("BusCount", "Bus Count")
plot_vehicle_vs_traffic("TruckCount", "Truck Count")
```

```python
# 8. Total Vehicle and Count Traffic Situation
# Group by 'Traffic Situation' and sum for each vehicle type
vehicle_types = ["CarCount", "BikeCount", "BusCount", "TruckCount"]

vehicle_totals = df.groupby('Traffic Situation')[vehicle_types].sum()

# Plot the results
vehicle_totals.plot(kind='bar', figsize=(10, 6), color=['skyblue', 'lightgreen', 'lightcoral', 'lightsalmon'])
plt.title('Total Vehicle Count by Traffic Situation for Each Vehicle Type')
plt.xlabel('Traffic Situation')
plt.ylabel('Total Vehicle Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend(title='Vehicle Types')
plt.show()
```

```python
# Sum the vehicle counts for each row (total vehicle count)
df['Total'] = df[vehicle_types].sum(axis=1)

# Group by 'Traffic Situation' and sum the total count for each traffic situation
traffic_totals = df.groupby('Traffic Situation')['Total'].sum()

# Plot the total vehicle count for each traffic situation
plt.figure(figsize=(8, 6))
traffic_totals.plot(kind='bar', color='coral')
plt.title('Total Vehicle Count by Traffic Situation')
plt.xlabel('Traffic Situation')
plt.ylabel('Total Vehicle Count')
plt.xticks(ticks=[0, 1, 2, 3], labels=['Heavy', 'High', 'Normal', 'Low'], rotation=0)  # Set custom x-axis labels
plt.show()
```

```python
# Task 9: Bar chart of total vehicle count per hour
plt.figure(figsize=(12, 6))
sns.barplot(x=df["Time"], y=df["Total"], estimator=sum, ci=None, palette="magma")
plt.title("Total Vehicle Count by Hour of the Day")
plt.xlabel("Hour of the Day")
plt.ylabel("Total Vehicles")
plt.show()


```
