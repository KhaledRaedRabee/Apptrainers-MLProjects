```python
df['Time'] = pd.to_datetime(df['Time'])
```

```python
# 1. Vehicle Distribution Analysis
vehicle_types = ['CarCount', 'BikeCount', 'BusCount', 'TruckCount']
plt.figure(figsize=(12, 6))
df[vehicle_types].plot(kind='box', vert=False)
plt.title('Distribution of Vehicle Counts')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(12, 6))
df[vehicle_types].hist(bins=20, alpha=0.7, layout=(2, 2), figsize=(12, 8))
plt.suptitle('Histogram of Vehicle Counts')
plt.show()
```

```python
# Task 2: Traffic Situation Distribution (Bar Chart)
plt.figure(figsize=(8, 5))
df['Traffic Situation'].value_counts().plot(kind='bar')
plt.title('Traffic Situation Distribution')
plt.xlabel('Traffic Situation')
plt.ylabel('Count')
plt.show()
```

```python
#Pie Chart
plt.figure(figsize=(8, 5))
df['Traffic Situation'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Traffic Situation Proportion')
plt.ylabel('')
plt.show()
```

```python
# Extract day of the week and hour for analysis
df['DayOfWeek'] = df['Time'].dt.day_name()
df['Hour'] = df['Time'].dt.hour
```

```python
# 3. Variation by Day of the Week
for vehicle in vehicle_types:
    plt.figure(figsize=(10, 6))
    sns.barplot(x='DayOfWeek', y=vehicle, data=df, estimator=sum, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], palette='Blues')
    plt.title(f'Total {vehicle} Count by Day of the Week')
    plt.xlabel('Day of the Week')
    plt.ylabel(f'Total {vehicle} Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# 4-7. Vehicle Count vs. Traffic Situation (Scatter & Violin Plots)
for vehicle in vehicle_types:
    plt.figure(figsize=(10, 6))
    sns.violinplot(x='Traffic Situation', y=vehicle, data=df)
    plt.title(f'{vehicle} vs. Traffic Situation')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# 8. Total Vehicle and Count Traffic Situation
# Group by 'Traffic Situation' and sum for each vehicle type
vehicle_totals = df.groupby('Traffic Situation')[vehicle_types].sum()

# Plot the results
vehicle_totals.plot(kind='bar', figsize=(10, 6), color=['skyblue', 'lightgreen', 'lightcoral', 'lightsalmon'])
plt.title('Total Vehicle Count by Traffic Situation for Each Vehicle Type')
plt.xlabel('Traffic Situation')
plt.ylabel('Total Vehicle Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend(title='Vehicle Types')
plt.show()
```

```python
# Sum the vehicle counts for each row (total vehicle count)
df['Total'] = df[vehicle_types].sum(axis=1)

# Group by 'Traffic Situation' and sum the total count for each traffic situation
traffic_totals = df.groupby('Traffic Situation')['Total'].sum()

# Plot the total vehicle count for each traffic situation
plt.figure(figsize=(8, 6))
traffic_totals.plot(kind='bar', color='coral')
plt.title('Total Vehicle Count by Traffic Situation')
plt.xlabel('Traffic Situation')
plt.ylabel('Total Vehicle Count')
plt.xticks(ticks=[0, 1, 2, 3], labels=['Heavy', 'High', 'Normal', 'Low'], rotation=0)  # Set custom x-axis labels
plt.show()
```

```python
# Ensure the 'Time' column is in datetime format
df['Time'] = pd.to_datetime(df['Time'])
# Extract the hour from the 'Time' column
df['Hour'] = df['Time'].dt.hour
```

```python
# Group by hour and sum the vehicle counts for each hour
hourly_totals = df.groupby('Hour')[['CarCount', 'BikeCount', 'BusCount', 'TruckCount']].sum()
# Create a new column for the total vehicle count per hour
hourly_totals['Total'] = hourly_totals.sum(axis=1)
```

```python
# Plot the total vehicle count by hour as a bar chart
plt.figure(figsize=(10, 6))
hourly_totals['Total'].plot(kind='bar', color='coral')
plt.title('Total Vehicle Count by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Total Vehicle Count')
plt.xticks(rotation=0)
plt.show()
```
